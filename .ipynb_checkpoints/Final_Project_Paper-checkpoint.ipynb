```python
# Load Census Bureau Household Pulse Survey data
hps = pd.read_csv("hps_data.csv")
# Load CDC data
cdc = pd.read_csv("COVID-19_Vaccinations_in_the_United_States_Jurisdiction.csv")
# Subset CDC data
cdc_new = cdc[cdc["Date"] == '7/21/2021']
```

```python
hps = hps.assign(state = hps.EST_ST)
cdc_new = cdc_new.assign(state = cdc_new.Location)

# Replace values in hps using a dictionary created from value of states in hps codebook
states_key = {1:'AL', 2:'AK', 4:'AZ', 5:'AR', 6:'CA', 8:'CO', 9:'CT', 10:'DE', 11:'DC', 12:'FL', 13:'GA', 15:'HI',16:'ID', 
              17:'IL', 18:'IN', 19:'IA', 20:'KS', 21:'KY', 22:'LA', 23:'ME', 24:'MD', 25:'MA', 26:'MI', 27:'MN', 28:'MS',
              29:'MO', 30:'MT', 31:'NE', 32:'NV', 33:'NH', 34:'NJ', 35:'NM', 36:'NY', 37:'NC', 38:'ND', 39:'OH', 40:'OK',
              41:'OR', 42:'PA', 44:'RI', 45:'SC', 46:'SD', 47:'TN', 48:'TX', 49:'UT', 50:'VT', 51:'VA', 53:'WA', 54:'WV',
              55:'WI', 56:'WY'}

hps.state.replace(states_key, inplace=True)
```

```python
# Choose variables needed from cdc_new
cdc_vars = cdc_new[['state','Distributed','Dist_Per_100K', 'Administered', 'Admin_Per_100K', 'Series_Complete_Pop_Pct', 
                    'Administered_Dose1_Pop_Pct']]

# Merge dataframes together
combined_df = hps.merge(cdc_vars, how="left", on="state")

# Replace missing data (-88 and -99 according to codebook) as NaN type
combined_df = combined_df.replace(-88, np.NaN) # Missing response to question
combined_df = combined_df.replace(-99, np.NaN) # Question seen but not answered

combined_df.RECVDVACC = combined_df.RECVDVACC.replace(2, 0) # Replacing "No" responses with 0 so this variable becomes a dummy

combined_df.WHYNORV1=combined_df.WHYNORV1.fillna(0) # side effects concern
combined_df.WHYNORV2=combined_df.WHYNORV2.fillna(0) # unsure of vaccine protection
combined_df.WHYNORV3=combined_df.WHYNORV3.fillna(0) # doesn't believe vaccine is necessary
combined_df.WHYNORV6=combined_df.WHYNORV6.fillna(0) # cost concern
combined_df.WHYNORV7=combined_df.WHYNORV7.fillna(0) # distrust of the vaccine
combined_df.WHYNORV8=combined_df.WHYNORV8.fillna(0) # distrust of the government
combined_df.WHYNORV9=combined_df.WHYNORV9.fillna(0) # doesn't see COVID-19 as a threat
combined_df.WHYNORV11=combined_df.WHYNORV11.fillna(0) # believes that one dose is enough protection
```

```python
combined_df = combined_df[['Series_Complete_Pop_Pct','RECVDVACC','GETVACRV','WHYNORV1','WHYNORV2','WHYNORV3','WHYNORV6','WHYNORV7','WHYNORV8','WHYNORV9','state']]
rs = np.random.RandomState(0)

combined_df_subset = combined_df[combined_df.RECVDVACC == 0]
combined_df_subset = combined_df_subset.drop('RECVDVACC', axis=1)
corr = combined_df_subset.corr()
corr_matrix = corr.style.background_gradient(cmap='coolwarm')

#miss_matrix = miss.matrix(combined_df_subset)
```

```python
combined_df_subset['GETVACRV_bins'] = pd.cut(combined_df_subset.GETVACRV, bins = 2, labels = ['1-3', '4-5'])
combined_df_subset = pd.get_dummies(combined_df_subset)
```

```python
y = combined_df_subset['GETVACRV_bins_4-5']
X = combined_df_subset[['Series_Complete_Pop_Pct', 'WHYNORV1', 'WHYNORV2', 'WHYNORV3', 'WHYNORV6', 'WHYNORV7', 'WHYNORV8',
       'WHYNORV9']]
train_X, test_X, train_y, test_y = train_test_split(X,y,test_size=.2,random_state=42)
```

## Introduction

The aim of this project is to deduce the significance of reasons why some people in the United States have not chosen to obtain a COVID-19 vaccine, particularly the significance of an echo chamber effect. This will be executed by analyzing survey responses and state level COVID-19 vaccination data relative to likelihood of survey respondents to get the COVID-19 vaccine. This report will cover background information on the problem, outline the data collected and used, describe methods of analysis, and discuss results and potential future analysis.


## Problem Statement and Background

Vaccine hesitancy is one of the primary reasons why the COVID-19 pandemic has persisted with such vigor for so long. Despite the COVID-19 vaccine first becoming available in the United States in December 2020, *Our World in Data* from the University of Oxford estimates that only 61% of the population in the United States is fully vaccinated against COVID-19 and that 72% of the population has had at least one dose of the COVID-19 vaccine (Richie et al., 2020). Along with the very gradual growth in recent vaccinations, the United States is very far off from reaching the “fully vaccinated” goal of 80% set by the Biden administration (Mayo Foundation for Medical Education and Research).

There is a myriad of reasons why people choose to not get vaccinated. One of the biggest reasons is distrust, either in medicine or the government. Previous research from the Queen Mary University of London has shown a correlation of populism and vaccine hesitancy, concluding that the two are driven by similar dynamics, a profound distrust in elites and experts (Khazan, 2021).  Aside from distrust, there are other factors to consider, such as misinformation and peer influence, both of which can potentially contribute to distrust.

Peer influence is especially one of the more interesting things to consider when contemplating the overall role that institutional distrust plays in vaccine hesitancy. An echo chamber effect is the idea that when we surround ourselves with people who share the same opinions, misinformation can spread and become pervasive and exacerbate people’s inability to consume information. In fact, in a network analysis studying the effects of misinformation on social media, the author found that a combination between opinion and network polarization resulted “in the synergetic effect that increases the virality of narratives that resonate with the echo chamber, meaning that the simple clustering together of users with a deviant world-view is enough to affect the virality of information items that resonate with their perspective” (Tornberg, 2018). 

In this analysis, I intend to control for vaccination rates by state as a measure of peer influence and echo chambers, hypothesizing that there is a negative relationship between vaccination rates and not being likely to become vaccinated against COVID-19. While this hypothesis is included in the purpose of this analysis, the other purpose of this analysis is to see what role factors of distrust play as well.


## Data

There were two datasets used in this project. The first dataset is from phase 3.2 of the Census Bureau’s Household Pulse Survey and collected on July 21st, 2021. This dataset contains survey responses from 64,562 participants from across the United States about how their lives have been impacted by the COVID-19 pandemic, including self-reported data on geographic location, vaccination status, indicators of vaccine hesitancy, and likelihood of becoming vaccinated, all of which were used in this analysis.

The second dataset is the Vaccination Rates in the United States by Jurisdiction data from the Centers for Disease Control and Prevention. This dataset contains vaccination rates in the United States and its territories organized according to state and date level observations. This dataset contains 21,912 observations collected between December 23rd, 2020 and November 2nd, 2021.

Prior to merging these datasets, I isolated a subset of the CDC dataset to only contain observations from July 21st, 2021, dropping the number of observations from 21,912 to 65. I then made a dictionary so that the **state** variables between the CDC and HPS datasets would be compatible. Using the dictionary, I then merged the two datasets on the **state** variables so that the vaccination rate of the respondents’ states could be included in the analysis. Lastly, I isolated a subset of the dataset again to restrict the observations to only include respondents who reported that they had not been vaccinated against COVID-19. All of these transformations were done using functions from the pandas package in python.

Our unit of analysis is a variable labeled **GETVACRV_bins_4-5** in the analysis. This variable was transformed from a categorical variable **GETVACRV** from the HPS dataset, which measured the likelihood of unvaccinated survey respondents becoming vaccinated against COVID-19 on a scale of 1-5, with 1 being “will definitely get a vaccine”, 2 being “will probably get a vaccine”, 3 being “unsure about getting a vaccine”, 4 being “will probably not get a vaccine”, and 5 being “will definitely not get a vaccine”. Logistically, I thought it would be easier to interpret this variable as a dichotomous variable, so I transformed it by binning it so that respondents who reported that they are unlikely to get vaccinated would be separated from those likely to get or unsure of getting vaccinated. I then further transformed it into a dummy variable, so that it equals 1 when a respondent reports that they’re unlikely to become vaccinated against COVID-19.

Our independent variables for the analysis are:

+ **Series_Complete_Pop_Pct**
+ **WHYNORV1**
+ **WHYNORV2**
+ **WHYNORV3**
+ **WHYNORV6**
+ **WHYNORV7**
+ **WHYNORV8**
+ **WHYNORV9**

**Series_Complete_Pop_Pct** is a continuous variable on scale of 0-100 that contains the percentage of fully vaccinated people for each state on July 21st, 2021. The remaining variables are dichotomous dummy variables and verbalize reasons why unvaccinated survey respondents are hesitant to get vaccinated against COVID-19; **WHYNORV1** concerns worries over potential side effects, **WHYNORV2** concerns worries over the strength of the vaccine’s protection, **WHYNORV3** concerns a belief that that the COVID-19 vaccine is unnecessary, **WHYNORV6** concerns worries over cost, **WHYNORV7** concerns distrust of the vaccine, **WHYNORV8** concerns distrust of the government, and **WHYNORV9** concerns a belief that COVID-19 does not pose a large threat.


After I finished merging my dataset, I needed to transform several of my variables due to the nature of survey data being coded in a certain way. For instance, in the HPS data, missing data or data that was not reported was coded as -88 and questions that were seen but did not receive a response were coded as -99. So, all of those values had to be re-coded as missing. In addition, I transformed my unit of analysis into a dichotomous dummy variable from a categorical variable so it would be better suited for a logistic regression and a naïve Bayesian estimator.

### Modeling Pipeline

For the statistical learning aspect of this model, I randomly separated the final dataframe into two datasets, one for training data, which contains 80% of the data, and the other for test data, which contains the remaining 20%. I then created a modeling pipeline to predict how a model that fits the training data will fit the test data. A modeling pipeline is essentially a collection of different models with various parameters that can be scaled and tuned to find the best model to fit the data according to a single criterion. 

For my modeling pipeline, I chose to include 6 different models. The models I chose to include are:

+ Linear regression
+ Logistic regression
+ Naïve Bayes Classifier
+ K Nearest Neighbors model
+ Decision Tree model
+ Random Forest model

All of these models are theoretically suitable for the kind of analysis I intend to execute using a dichotomous outcome variable. Of these models, only K Nearest Neighbors, Decision Tree, and Random Forest have parameters that needed to be tuned. For K Nearest Neighbors, I chose to tune the n_neighbors parameter, for the Decision Tree, I chose to tune max_depth, and for Random Forest, I chose to tune n_estimators and max_depth. Of the remaining models, I also had a sense that the Naïve Bayes Classifier and the Logistic Regression would provide good fits for the dataset because are both intended to be used with binary variables. The Linear Regression I was less sure of because there are domain issues with using it, but I still included it because I thought it would still be a good measure and because it’s easier to interpret than the Logistic Regression.

The criterion I ultimately decided to use to measure which model is the best is the roc_auc score. This score is often used to determine accuracy in models with dichotomous outcome variables by calculating the area under an ROC curve by calculating if a model correctly classified positive outcomes.

In addition to these tools, I used the sklearn, plotnine, matplotlib, pandas, and numpy packages to assist my project.

## Results

```python
fold_generator = KFold(n_splits=5, shuffle=True,random_state=42)

# Specify the preprocessing steps
preprocess = ColumnTransformer(transformers=[('num', pp.MinMaxScaler(), ['Series_Complete_Pop_Pct',
       'WHYNORV1', 'WHYNORV2', 'WHYNORV3', 'WHYNORV6', 'WHYNORV7', 'WHYNORV8',
       'WHYNORV9'])])


# Create our model pipe
pipe = Pipeline(steps=[('pre_process', preprocess),
                       ('model',None)])


# Specify the models and their repsective tuning parameters. 
search_space = [
    # Linear Regression
    {'model' : [LM()]},
    
    # Logistic Regression
    {'model' : [Logit()]},
    
    # Naive Bayes Classifier
    {'model' : [GaussianNB()]},
    
    # KNN with K tuning param
    {'model' : [KNN()],
     'model__n_neighbors':[10,15,20,25,30]},
    
    # Decision Tree with the Max Depth Param
    {'model': [DTree()],
     'model__max_depth':[1,2,3,4,5]},
    
    # Random forest with the N Estimators tuning param
    {'model' : [RF()],
     'model__max_depth':[1,2,3],
     'model__n_estimators':[500,1000,1250,1500]},
]

search_roc = GridSearchCV(pipe, search_space, 
                      cv = fold_generator,
                      scoring='roc_auc',
                      n_jobs=4)

search_roc_fit = search_roc.fit(train_X, train_y)
```

```python
best_model = search_roc.best_params_
score = search_roc.best_score_
pred_y = search_roc.predict(test_X)
test_pred = m.roc_auc_score(test_y,pred_y)
model_dict = {'Best_model':best_model, 'Model_score_train':score, 'Model_score_test':test_pred}
model_df = pd.DataFrame(data=model_dict)
```

Here are the results of which model performed best and what its roc_auc score is:

```python
model_df
```

Of all the potential models that were featured in the pipeline, the model that proved to be the best fit was the logistic regression with roc_auc score of 0.756, indicating that this model correctly classified observations 75.6% of the time for the training dataset. However, for the test dataset, the model proved to be slightly less accurate and correctly classified observations 71.9% of the time. While these numbers could be improved, they are still fairly accurate.

Now, let’s look at how the different independent variables compare in terms of their relative importance:

```python
rf_mod = search_roc.best_estimator_
vi = permutation_importance(rf_mod,train_X,train_y,n_repeats=30)
vi_dat = pd.DataFrame(dict(variable=train_X.columns,
                           vi = vi['importances_mean'],
                           std = vi['importances_std']))

vi_dat['low'] = vi_dat['vi'] - 2*vi_dat['std']
vi_dat['high'] = vi_dat['vi'] + 2*vi_dat['std']
vi_dat = vi_dat.sort_values(by="vi",ascending=False).reset_index(drop=True)
vi_dat
```

```python
(
    ggplot(vi_dat,
          aes(x="variable",y="vi")) +
    geom_col(alpha=.5) +
    geom_point() +
    geom_errorbar(aes(ymin="low",ymax="high"),width=.2) +
    theme_bw() +
    scale_x_discrete(limits=vi_dat.variable.tolist()) +
    coord_flip() +
    labs(y="Reduction in AUC ROC score",x="")
)
```

As you can see from the table and boxplot above, the variables with the most relative importance are **WHYNORV7**, **WHYNORV8**, and **WHYNORV3**. These are the variables that concern distrust of the vaccine, distrust in the government, and a belief that the COVID-19 vaccine is necessary. Unfortunately, the variable that we hypothesized was going to have a substantive effect on the model, **Series_Complete_Pop_Pct**, ended up having one of the least substantive effects on the likelihood of becoming vaccinated against COVID-19.
